```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,4*np.pi,0.1) #0~4πの間、0.1の間隔でxを変化させる
y = np.sin(x) #xに対してsin(x)を求める

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
#最近の環境ではfig.show()いらない?(https://qiita.com/7of9/items/f9060a686ae1c45acae8)
```
