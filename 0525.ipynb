```python
# 教科書の例

from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def system(y,t):
    if t < 10.0:
        u = 0.0
    else:
        u = 1.0
    dydt = (-y + u) / 5.0
    return dydt

y0=0.5
t = np.arange(0,40,0.0625)
y = odeint(system,y0,t)

fig, ax = plt.subplots()
ax.plot(t,y,label='y')
ax.plot(t,1*(t >= 10), ls='--', label='u')
ax.set_xlabel('t')
ax.set_ylabel('u,y')
ax.legend(loc='best')
ax.grid(ls=':')
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def system2(xv,t):
    x,v = xv
    dxvdt = [v,-5*v-4*x]
    return dxvdt

y0=[0,1]
t = np.arange(0,10,0.0625)
y = odeint(system2,y0,t)

fig, ax = plt.subplots()
ax.plot(t,y,label='y')
ax.set_xlabel('t')
ax.set_ylabel('x,v')
# ax.legend(loc='best')
ax.grid(ls=':')
```
